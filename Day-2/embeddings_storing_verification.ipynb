Import Modules

```python
from utils.embedding import generate_embedding
from pymongo import MongoClient
from utils.config import MONGO_URI, DATABASE_NAME, COLLECTION_NAME
```

Connection to MongoDB

```python
client = MongoClient(MONGO_URI)
collection = client[DATABASE_NAME][COLLECTION_NAME]
```

Defining the Input

```python
symptoms = "acne on face with oily skin"
doctor_notes = "moderate inflammatory acne"
```

RUN FIRST TIME

```python
embedding1 = generate_embedding(symptoms, doctor_notes)
print("First run completed")
```

MongoDB checking and verification

```python
doc = collection.find_one({
    "symptoms": symptoms,
    "doctor_notes": doctor_notes
})

print("Found in DB:", doc is not None)
print("Has embedding:", "embedding" in doc)
```

From this we can able to see that. our provided input is converted into the embedding form and properly stoed in the MongoDB with it's respective embedding.

Check for duplicates if any

```python
count = collection.count_documents({
    "symptoms": symptoms,
    "doctor_notes": doctor_notes
})

print("Number of records:", count)
```
