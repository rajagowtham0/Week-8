```python
import matplotlib.pyplot as plt

# Data
data_size = [64, 65, 59, 55, 62, 60, 116, 55, 56, 69,
             60, 52, 50, 54, 66, 36, 105, 134, 155, 92]

retrieval_time = [0.0052, 0.0036, 0.0118, 0.0017, 0.0022, 0.0035,
                  0.0415, 0.0056, 0.0038, 0.0031, 0.0034, 0.004,
                  0.0037, 0.0036, 0.0972, 0.0664, 0.0971, 0.0959,
                  0.0731, 0.0624]

api_time = [0.0084, 0.0036, 0.0159, 0.0044, 0.0042, 0.0056,
            0.0424, 0.0082, 0.0038, 0.0058, 0.0056, 0.005,
            0.006, 0.0061, 0.1008, 0.0678, 0.0991, 0.0977,
            0.075, 0.064]

# X positions (unique spacing)
x_pos = list(range(len(data_size)))

plt.figure(figsize=(14, 6))

# Thick & clean lines
plt.plot(x_pos, retrieval_time, marker='o', linewidth=2.5, markersize=7,
         label='Retrieval Time')

plt.plot(x_pos, api_time, marker='s', linewidth=2.5, markersize=7,
         label='API Response Time')

# Annotation offsets (STRICT separation)
offset_up = 0.003
offset_down = 0.003

for i in range(len(x_pos)):
    # API → ALWAYS ABOVE
    plt.text(x_pos[i], api_time[i] + offset_up,
             f"{api_time[i]:.4f}",
             ha='center', va='bottom', fontsize=8)

    # Retrieval → ALWAYS BELOW
    plt.text(x_pos[i], retrieval_time[i] - offset_down,
             f"{retrieval_time[i]:.4f}",
             ha='center', va='top', fontsize=8)

# X-axis labels
plt.xticks(x_pos, data_size, rotation=45)

# Labels
plt.xlabel("Data Size (Bytes)", fontsize=12)
plt.ylabel("Time (seconds)", fontsize=12)
plt.title("Retrieval Time vs API Response Time (Per Sample)", fontsize=14)

# Remove grid
plt.grid(False)

# Clean axes (only left & bottom)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Legend
plt.legend()

plt.tight_layout()
plt.show()
```
