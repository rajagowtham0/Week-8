Import libraries

```python
import time
import pandas as pd

from retrieval.retrieval_engine import initialize_engine, analyze_case
from utils.embedding import combine_text
from retrieval.retrieval_engine import case_ids
```

Initialise system

```python
initialize_engine()

DATASET_SIZE = len(case_ids)
print("Dataset Size:", DATASET_SIZE)
```

Inputs and processing

```python
test_cases = [
    ("acne on face with oily skin", "moderate acne"),
    ("hair fall and thinning hair", "possible alopecia"),
    ("dark patches on skin", "hyperpigmentation")
]

results = []

for symptoms, notes in test_cases:

    combined_text = combine_text(symptoms, notes)

    # Measure API time
    api_start = time.time()

    # Measure retrieval time
    retrieval_start = time.time()
    output = analyze_case(combined_text)
    retrieval_time = round(time.time() - retrieval_start, 4)

    api_time = round(time.time() - api_start, 4)

    results.append({
        "input": symptoms,
        "dataset_size": DATASET_SIZE,
        "retrieval_time_sec": retrieval_time,
        "api_time_sec": api_time,
        "output": output
    })
```

```python
df = pd.DataFrame(results)

# Show only performance columns
df[["input", "dataset_size", "retrieval_time_sec", "api_time_sec"]]
```
